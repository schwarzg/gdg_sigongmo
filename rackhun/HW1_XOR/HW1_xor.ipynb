# Import library

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

# Parameters

```python
n_model = 3

hidden_units = 2
learning_rate = 0.1
n_epoch = 4000
batch_size = 4
```

# Data

```python
X        = np.array([ [0, 0], [0, 1], [1, 0], [1, 1] ])

Y_and = np.array([ [0], [0], [0], [1] ])
Y_or    = np.array([ [0], [1], [1], [1] ])
Y_xor  = np.array([ [0], [1], [1], [0] ])

X_dim = X.shape[1]
Y_dim = Y_and.shape[1]
```

# Model

```python
models=[]
for m in range(n_model):
    models.append(Sequential())

    # structure
    models[m].add(Dense(input_dim=X_dim, units=hidden_units))
    models[m].add(Activation('sigmoid'))
    #models[m].add(Activation('relu'))
    models[m].add(Dense(units=Y_dim))
    models[m].compile(loss='binary_crossentropy', optimizer=SGD(lr=learning_rate))
```

# Training

```python
models[0].fit(X, Y_and, epochs=n_epoch, batch_size=batch_size)
models[1].fit(X, Y_or, epochs=n_epoch, batch_size=batch_size)
models[2].fit(X, Y_xor, epochs=n_epoch, batch_size=batch_size)
```

# Result

```python
result_and = models[0].predict_classes(X, batch_size=batch_size)
result_or   = models[1].predict_classes(X, batch_size=batch_size)
result_xor = models[2].predict_classes(X, batch_size=batch_size)
```
