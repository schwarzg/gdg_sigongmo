```python
#   (Multi-layer Perceptron, MLP):       .
# pp.123 - 136
# XOR Gate - 3 layers model

import numpy as np
X = np.array([[0,0],[0,1],[1,0],[1,1]]) # 
Y = np.array([[0],[1],[1],[0]]) # 

from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD

#  - 
model = Sequential()
model.add(Dense(input_dim=2, units=2))
model.add(Activation('sigmoid'))

#  - 
model.add(Dense(units=1))
model.add(Activation('sigmoid'))

lr = 0.1 # learning rate
model.compile(loss='binary_crossentropy', optimizer=SGD(lr=lr))
model.fit(X,Y, epochs=2000, batch_size=4)

#prediction
classes = model.predict_classes(X, batch_size=4)
prob = model.predict_proba(X, batch_size=4)
print(' :')
print(Y==classes)
print()
print(' : ')
print(prob)
```
